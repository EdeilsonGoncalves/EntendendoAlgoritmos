```python
'''Suponha que voce tenha uma lista com 128 nomes e esteja fazendo uma pesquisa binaria.
qual seria o numero maximo de etapas que voce levaria para encontrar o nome desejado?'''


def pesquisa_binaria(lista, item):
  baixo = 0
  alto = len(lista) - 1

  while baixo <= alto:
    meio = (baixo + alto) // 2
    chute = lista[meio]
    if chute == item:
      return meio
    if chute > item:
      alto = meio - 1
    else:
      baixo = meio + 1
  return None


lista = ['Ana', 'Beatriz', 'Carlos', 'Daniel', 'Eduardo', 'Felipe', 'Guilherme', 'Heitor', 'Ian']

print(pesquisa_binaria(lista, 'Eduardo')) # => 4
```
