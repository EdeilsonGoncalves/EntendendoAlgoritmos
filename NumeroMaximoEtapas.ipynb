```python
#Suponha que voce duplique o tamanho da lista. Qual seria o número máximo de etapas agora?

import math

def calcular_max_etapas(num_elements):
    # Calcular o número máximo de etapas
    max_steps = math.ceil(math.log2(num_elements))

    # Imprimir o resultado
    print(f"O número máximo de etapas para encontrar um nome em uma lista de {num_elements} elementos é {max_steps}.")

# Exemplo com uma lista de 128 elementos
calcular_max_etapas(128)

# Exemplo com uma lista de 256 elementos (duplicado)
calcular_max_etapas(256)
```
