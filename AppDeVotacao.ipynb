```python
class Eleicao:
    def __init__(self):
        self.ministerios = []
        self.resultados = {}

    def adicionar_ministerio(self, ministerio):
        self.ministerios.append(ministerio)
        self.resultados[ministerio] = {}

    def votar(self, ministerio, nome):
        if ministerio in self.ministerios:
            if nome in self.resultados[ministerio]:
                self.resultados[ministerio][nome] += 1
            else:
                self.resultados[ministerio][nome] = 1

            print(f"Voto registrado para {nome} no ministério de {ministerio}.")
        else:
            print(f"Ministério '{ministerio}' não encontrado.")

    def listar_resultados(self):
        print("\nResultados:")  # Adiciona uma linha em branco antes de "Resultados:"
        for ministerio, votos in self.resultados.items():
            print(f"\n{ministerio}:")
            for nome, contagem in votos.items():
                plural_s = "s" if contagem > 1 else ""
                print(f"{nome}: {contagem} voto{plural_s}")
            print()  # Adiciona uma linha em branco entre ministérios

def main():
    eleicao = Eleicao()

    while True:
        print("\nOpções:")
        print("1. Adicionar Ministério")
        print("2. Votar")
        print("3. Listar Resultados")
        print("4. Sair")

        escolha = input("Escolha uma opção: ")

        if escolha == "1":
            ministerio = input("Digite o nome do ministério: ")
            eleicao.adicionar_ministerio(ministerio)
        elif escolha == "2":
            if eleicao.ministerios:
                ministerio = input("Digite o nome do ministério para votar: ")
                nome = input("Digite o seu nome para votar: ")
                eleicao.votar(ministerio, nome)
            else:
                print("Nenhum ministério cadastrado.")
        elif escolha == "3":
            eleicao.listar_resultados()
        elif escolha == "4":
            break
        else:
            print("Opção inválida. Tente novamente.")

if __name__ == "__main__":
    main()
```
